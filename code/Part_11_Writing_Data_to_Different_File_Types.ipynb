# Part 11: Writing and Saving Data on Disk

```python
import polars as pl
pl.Config.set_tbl_rows(10)
pl.Config.set_fmt_str_lengths(50)
df = pl.read_csv('data/survey_results_public.csv')
```

```python
# Filter for India and export to csv
india_df = df.filter(pl.col('Country') == 'India')
india_df.head()
```

```python
# Write to csv
india_df.write_csv('data/modified.csv')
```

```python
# Other types of files (Tab delimited)
india_df.write_csv('data/modified.tsv', separator='\t')
```

```python
# Write to parquet
india_df.write_parquet('data/modified.parquet')
```

```python
# Read the parquet file.
pl.read_parquet('data/modified.parquet').head()
```

```python
# Write to json
india_df.write_json('data/modified.json', row_oriented=True)
```
